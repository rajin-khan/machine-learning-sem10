# EDA(Exploratory Data Analysis)

## What is Exploratory Data Analysis (EDA)?

EDA, which stands for Exploratory Data Analysis, is an approach used in data science and statistics to analyze and summarize datasets. It involves investigating the data to understand its main characteristics, patterns, and relationships before applying more advanced modeling or analysis techniques.

## Goal of Exploratory Data Analysis (EDA):
The goal of Exploratory Data Analysis (EDA) is to understand and gain insights from the data at hand. EDA involves examining and visualizing the data using statistical and graphical techniques to uncover patterns, identify relationships, detect anomalies, and summarize the main characteristics of the dataset.

The specific objectives of EDA include:

**Data Familiarization:**

EDA helps in getting acquainted with the data by examining its structure, dimensions, and the types of variables present.

**Data Quality Check:**

EDA allows you to identify and handle missing values, outliers, and inconsistencies in the data. It helps in assessing the overall quality of the dataset.

**Variable Identification:**

EDA helps in understanding the nature of each variable in the dataset, whether it is categorical, numerical, discrete, continuous, or a combination of these.

**Univariate Analysis:**

EDA involves analyzing individual variables to determine their distribution, central tendency (mean, median), spread (variance, standard deviation), and identifying any patterns or anomalies.

**Bivariate and Multivariate Analysis:**

EDA explores relationships between variables, identifying correlations, associations, and dependencies among them. It helps in understanding how variables interact with each other and how they collectively contribute to the overall patterns in the data.

**Feature Selection:**

EDA can assist in selecting the most relevant features or variables for subsequent modeling or analysis tasks. It helps in identifying which variables are most informative and influential for the desired outcome.

**Communication and Reporting:**

EDA provides a visual and descriptive summary of the data, making it easier to communicate findings and insights to stakeholders. It aids in making data-driven decisions and formulating hypotheses for further analysis.

EDA, data analysts and scientists can gain a deeper understanding of the dataset, uncover patterns and insights, validate assumptions, and guide subsequent data preprocessing, feature engineering, and modeling steps.

## STEPS to perform Exploratory Data Analysis (EDA) : 

EDA involves examining the data visually and statistically to discover patterns, identify anomalies, and test assumptions.

Here are some common techniques and steps involved in EDA:

**Data collection:**

Gather the relevant dataset for analysis, which may come from various sources such as databases, spreadsheets, or online repositories.

**Data cleaning:**

Preprocess and clean the data to handle missing values, outliers, and inconsistencies. This step involves tasks like removing duplicates, imputing missing values, and correcting data format issues.

**Data visualization:**

Create visual representations of the data using graphs, charts, and plots. Visualizing the data helps identify patterns, trends, and relationships between variables.

**Summary statistics:** 

Calculate descriptive statistics such as mean, median, standard deviation, and quartiles to understand the central tendency, dispersion, and shape of the data distribution.

**Data transformation:** 

Apply transformations to the data, such as scaling, normalization, or log transformations, to make it suitable for further analysis or modeling.

**Feature engineering:** 

Create new features or derive meaningful information from existing features to enhance the predictive power of the data. This step involves techniques like encoding categorical variables, creating interaction terms, or extracting relevant information from text or timestamps.

**Correlation analysis:** 

Explore the relationships between variables by calculating correlation coefficients. This helps identify strong or weak associations and potential multicollinearity issues.

**Outlier detection:** 

Identify and handle outliers that can affect the analysis or modeling results. Outliers are data points that significantly deviate from the rest of the data.

**Hypothesis testing:** 

Formulate hypotheses and perform statistical tests to validate or reject them. This step helps make inferences about the data and draw conclusions.

**Iterative analysis:** 

Repeat the above steps as necessary, focusing on different aspects of the data or exploring specific hypotheses. EDA is an iterative process that often involves refining the analysis based on the insights gained along the way.

By conducting EDA, data scientists and analysts can gain a deeper understanding of the data, uncover patterns or relationships, detect anomalies, and generate insights to inform decision-making or guide further analysis or modeling tasks.

Handling attributes with **`null`** values in machine learning depends on the specific context and the amount of missing data. Here are a few best practices for dealing with **`null`** values:

- **Analyze the Missing Data:** Start by analyzing the extent and pattern of missing data in the attribute. Understand the reasons for the missingness and assess if it is Missing Completely at Random (MCAR), Missing at Random (MAR), or Missing Not at Random (MNAR). This analysis can guide the appropriate handling strategy.

- **Imputation with Mean/Median/Mode:** If the missingness is random or does not introduce bias, you can consider imputing the missing values. For numerical attributes, impute with the mean or median value of the non-missing values. Median is generally preferred if there are outliers, as it is more robust to extreme values. For categorical attributes, impute with the mode (most frequent value).

- **Imputation with Regression Models:** If the missingness has a pattern and can be predicted from other attributes, you can use regression models to impute the missing values. Fit a regression model using the non-missing values as predictors and the attribute with missing values as the target. Use the fitted model to predict the missing values.

- **Multiple Imputation:** If you have a substantial amount of missing data, consider using multiple imputation techniques. Multiple imputation creates several imputed datasets based on a statistical model, allowing for the estimation of missing values' uncertainty and preserving variability.

- **Drop the Column or Rows:** If the attribute has a high percentage of missing values or is not considered essential for the analysis, you may choose to drop the column altogether. Similarly, if a particular instance has multiple missing values across important attributes, you may consider removing that specific row from the dataset.

Ultimately, the best approach depends on the specific dataset, the domain, and the goals of the machine learning task. It's crucial to assess the impact of different strategies on the model's performance and consider the potential biases introduced by imputation or data removal.

#### **`Categorical and Numerical Data:`** 

Categorical data and numerical data are two types of data variables that are commonly encountered in datasets. They differ in the nature of the information they represent and the types of analysis and visualizations that can be applied to them.

**Categorical Data:** 
Categorical data represents variables that have distinct categories or groups. These categories are usually represented by labels or names. Categorical variables can be further classified into nominal and ordinal variables.

`1. Nominal Variables:`

Nominal variables represent categories that have no inherent order or ranking. Examples include gender (e.g., male, female), color (e.g., red, blue, green), or country (e.g., USA, Canada, Australia). Nominal variables can be encoded using one-hot encoding or label encoding for analysis.

`2. Ordinal Variables:`

Ordinal variables represent categories that have a specific order or ranking. The categories have a relative value or hierarchy. Examples include education level (e.g., high school, bachelor's, master's, Ph.D.), satisfaction level (e.g., low, medium, high), or rating (e.g., 1-star, 2-star, 3-star). Ordinal variables can be encoded with numeric values to represent their relative order.

For categorical data, common techniques for analysis and visualization include count plots, bar charts, pie charts, and cross-tabulations.

**Numerical Data:**
Numerical data represents variables that contain numeric values and can be further classified into two subtypes:

`1. Discrete Variables:`

Discrete variables represent whole numbers or integers that have a countable number of possible values. Examples include the number of siblings, number of items sold, or the count of a particular event. Discrete variables often have a finite range of values.

`2. Continuous Variables:`

Continuous variables represent numeric quantities that can take any value within a range or interval. Examples include age, height, weight, temperature, or income. Continuous variables can have an infinite number of possible values within a range.

For numerical data, common techniques for analysis and visualization include histograms, box plots, scatter plots, line plots, and correlation matrices.

Understanding the distinction between categorical and numerical data is important because it guides the choice of appropriate analysis techniques and visualization methods to extract meaningful insights from the data.

### Correlation

**`Correlation`** is a statistical measure that indicates the extent to which two or more variables **fluctuate** in relation to each other. A **positive correlation** indicates the extent to which those variables **increase or decrease** in parallel; a **negative correlation** indicates the extent to which one variable **increases as the other decreases**.

A correlation coefficient is a statistical measure, of the degree to which changes to the value of one variable predict change to the value of another. 

To create a **`correlation`** for our dataset, we first need to encode all the **`categorical`** attributes of our dataframe.

````python
from sklearn.preprocessing import LabelEncoder

# Create an instance of LabelEncoder
label_encoder = LabelEncoder() 

# Encode the categorical column
df['Name'] = label_encoder.fit_transform(df['Name'])
df['Sex'] = label_encoder.fit_transform(df['Sex'])
df['Embarked'] = label_encoder.fit_transform(df['Embarked'])
df['Ticket'] = label_encoder.fit_transform(df['Ticket'])
````

````python
# dataset attribute correlation
df.corr()
````

````python
#Correlation plot
sns.set(rc = {'figure.figsize':(24, 16)})
sns.heatmap(df.corr(), annot=True, cmap="YlGnBu")
````

````python
sns.set(rc = {'figure.figsize':(21, 13)})
sns.set(font_scale = 1.67)
sns.barplot(df.corr(), errorbar=None, orient="v")
````
